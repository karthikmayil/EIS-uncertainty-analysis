```python
import numpy as np
import proplot as pplot
import pandas as pd
from scipy import interpolate
import seaborn as sns
```

```python
# read in raw phase/magnitude extracted from WebplotDigitizer
def read_spectra(i):
    data = pd.read_csv(f'wang_2020_EIS_spectra/raw_spectra/{i}.csv')
    phase_data = data.iloc[1:,0:2].dropna().reset_index(drop=True).astype(float)
    phase_data.columns=['f','phase']
    phase_data = phase_data.sort_values(by=['f'])
    Z_data = data.iloc[1:,2:4].dropna().reset_index(drop=True).astype(float)
    Z_data.columns=['f','Z']
    Z_data = Z_data.sort_values(by=['f'])
    return phase_data,Z_data
```

```python
# map extracted phase/magnitude data to a evenly spaced set of frequencies, convert to Re and Im
def interp_spectra(phase_data,Z_data,axs=None,plot=False):
    frange = np.logspace(np.log10(np.max([phase_data.f.min(),Z_data.f.min()])),np.log10(np.min([phase_data.f.max(),Z_data.f.max()])),100)

    phase_func = interpolate.interp1d(phase_data.f,phase_data.phase,bounds_error=False,fill_value='extrapolate')
    phase_interp = phase_func(frange)

    Z_func = interpolate.interp1d(Z_data.f,Z_data.Z,bounds_error=False,fill_value='extrapolate')
    Z_interp = Z_func(frange)
    
    Z_real = Z_interp*np.cos(-1*phase_interp*np.pi/180)
    Z_imag = Z_interp*np.sin(-1*phase_interp*np.pi/180)
    
    interpolated_spectra=pd.DataFrame({'f':frange,'Re':Z_real,'Im':Z_imag,'Mag':Z_interp,'Phase':phase_interp})
    
    
    if plot:
        ax = axs[0]
        ax.format(xlabel='f $Hz$', ylabel=r'Phase $\degree$')#,xlocator='log', xminorlocator='logminor')
        ax.semilogx(phase_data.f,phase_data.phase,'ko',mfc='w',label='WpD Extraction')
        ax.semilogx(frange,phase_interp,'-k',label='Interpolated Curve')
        
        ax2 = ax.twinx()
        ax2.format(ycolor='red', ylabel=r'|Z| $Ohm$')
        ax2.semilogx(Z_data.f,Z_data.Z,'or',mfc='w',label='WpD Extraction')
        
        ax2.semilogx(frange,Z_interp,'-r',label='Interpolated Curve')

        # ax.legend(loc='bottom')#loc='ur')

        ax = axs[1]
        
        ax.semilogx(frange,interpolated_spectra.Re,'-or',mfc='w',mew=0.5,label='Re')
        ax.semilogx(frange,-interpolated_spectra.Im,'-or',mew=0.5,label='-Im')
        ax.format(xlabel='f',ylabel='(Ohms)')
        
        ax = axs[2]
        ax.plot(interpolated_spectra.Re,-1*interpolated_spectra.Im,'-or',mfc='w',label='exp')
        ax.format(xlabel='Re(Z) (Ohms)',ylabel='Im(Z) (Ohms)',xlim=[0,Z_real.max()*1.1])

        return interpolated_spectra,axs
    else:
        return interpolated_spectra
```

```python
# cell gemoetry specified in experimental section
diam = 1.27 #cm
area = np.pi* (diam/2)**2


def Rion_to_tau(Rion,Lum,porosity):
    
    # bulk conductivity of 10mM TABClO4, blocking electrolyte used in experiment. Taken from paper.
    conductivity0 = 0.3*1e-3 #S/cm  
    L = Lum*1e-4 #cm
    
    return Rion*area*conductivity0*porosity/(2*L)

def Rion_to_keff(Rion,Lum,porosity):
    
    return Rion*area/(2*L)
```

```python
from impedance.models.circuits import CustomCircuit
```

```python
var_names = ['R_HRF','R_cc','Q_cc','gamma_cc','R_loc','Q_loc','gamma_loc']

circuit_shorthand = 'R_0-p(R_1,CPE_1)-TLMQ'
lbs = [0,50,1e-6,0.5,50,0.1e-3,0.5]
ubs = [100,200,50e-6,1,500,10e-3,1]

customCircuit2 = CustomCircuit(initial_guess=np.mean([lbs,ubs],axis=0),
                                  circuit=circuit_shorthand)



def fit_GTLM(spectra):

    Zexp = np.array([complex(spectra.Re.values[i],spectra.Im.values[i]) for i in range(len(spectra))])

    customCircuit2.fit(frequencies=spectra.f.values, impedance=Zexp, 
                       sigma = np.hstack([5*np.ones(len(spectra)), 4*np.ones(len(spectra))]), #0.01*np.hstack([spectra.Re, spectra.Im]),
                       bounds=(lbs,ubs))
    
    return customCircuit2.parameters_,customCircuit2.conf_
```

```python
electrode_specs = pd.read_excel('wang_electrodes.xlsx',sheet_name='specs')
electrode_specs = electrode_specs.dropna().sort_values(by='figure').reset_index(drop=True)
electrode_specs
```

```python
# samples = np.arange(1,5)

fig, axx = pplot.subplots(ncols=3,nrows=len(electrode_specs), 
                          aspect=1, width=9, 
                          span=False, sharex=False,sharey=False,wratios=(2,2,1),
                          bottom='5em', right='5em',)  # margin spacing overrides
axx.format(abc=True, abcloc='l',
           rowlabels=list(electrode_specs['electrode'].values),
           collabels=['Phase + Magnitude', 'Re + IM Impedance', 'Nyquist']
)
electrode_specs['mu_tau'] = np.zeros(len(electrode_specs))
electrode_specs['std_tau'] = np.zeros(len(electrode_specs))
STLM_fit_params = []
for i in electrode_specs.index:
    phase_data,Z_data = read_spectra(electrode_specs.loc[i,'figure'])
    interpolated_spectra,axs = interp_spectra(phase_data,Z_data,axs=axx[i,:],plot=True)
    interpolated_spectra.to_csv(f'wang_2020_EIS_spectra/processed_spectra/A{i}.csv',sep=',',index=False)

    mu,sigma = fit_GTLM(interpolated_spectra)
    tau = Rion_to_tau(mu[4],electrode_specs.loc[i,'thickness'],electrode_specs.loc[i,'porosity_calc'])
    print(electrode_specs.loc[i,'electrode'],mu[4],sigma[4],tau,tau*sigma[4]/mu[4])
    
    customCircuit2.parameters_ = mu
    Zpred = customCircuit2.predict(interpolated_spectra.f.values)
    axx[i,2].plot(np.real(Zpred),-np.imag(Zpred),'-b',label='STLM fit')
    axx[i,2].format(xlabel='Re(Z) (Ohms)',ylabel='Im(Z) (Ohms)',xlim=[0,np.max(axx[i,2].get_xlim())])
    
    axx[i,1].plot(interpolated_spectra.f,np.real(Zpred),'--b',label='STLM fit, Re')
    axx[i,1].plot(interpolated_spectra.f,-np.imag(Zpred),'-b',label='STLM fit, -Im')
    
    electrode_specs.loc[i,'mu_tau'] = tau
    electrode_specs.loc[i,'std_tau'] = tau*sigma[4]/mu[4]
    
    STLM_fit_params.append(list(mu)+['mean',electrode_specs.loc[i,'electrode']])
    STLM_fit_params.append(list(sigma)+['std',electrode_specs.loc[i,'electrode']])
    
axx.legend(ncols=1,fontsize=7);
STLM_fit_results = pd.DataFrame(STLM_fit_params,columns=var_names+['metric','electrode'])
```
